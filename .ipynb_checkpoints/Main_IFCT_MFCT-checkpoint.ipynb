```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.ticker as mticker
import pandas as pd
import seaborn as sns
import numpy as np
from scipy.optimize import curve_fit
from uncertainties import ufloat
from lmfit import Model
import statistics as stat
from scipy.stats.distributions import  t
from matplotlib.offsetbox import (TextArea, DrawingArea, OffsetImage,
                                  AnnotationBbox)
import matplotlib.font_manager
from matplotlib.lines import Line2D
from matplotlib.font_manager import FontProperties

import matplotlib.font_manager
matplotlib.font_manager.findSystemFonts(fontpaths=None, fontext='ttf')
matplotlib.rcParams['font.serif'] = ['Times']
matplotlib.rcParams['font.family'] = "DejaVu Sans"
fontsize = 30
```

```python
# n1*M^n2*log(n3M)
def objective(x, n1, n2, n3):
    return n1 + n2*x + np.log10(n3 + (x*np.log(10)))
```

```python
# n1*M^n2
def objective_withoutLog(x, n1, n2):
    return n1 + n2*x
```

```python
# load the dataset for MFCT and IFCT: Localized (DC ~ M^0) and Systemic (DC ~ M^0.5) infection 
# LN vol scaling as M^1/2

# Motion: Brownian; 
# dataMC = pd.read_csv('Data/BM_MFCT_DCConstant.csv', header = None, delimiter='\t')
# dataMH = pd.read_csv('Data/BM_MFCT_DCHalf.csv', header = None, delimiter='\t')
# dataFC = pd.read_csv('Data/BM_IFCT_DCConstant.csv', header = None, delimiter='\t')
# dataFH = pd.read_csv('Data/BM_IFCT_DCHalf.csv', header = None,delimiter='\t')

# Motion: CRW
dataMC = pd.read_csv('Data/CRW_MFCT_DCConstant.csv', header = None, delimiter='\t')
dataMH = pd.read_csv('Data/CRW_MFCT_DCHalf.csv', header = None, delimiter='\t')
dataFC = pd.read_csv('Data/CRW_IFCT_DCConstant.csv', header = None, delimiter='\t')
dataFH = pd.read_csv('Data/CRW_IFCT_DCHalf.csv', header = None,delimiter='\t')

data_logMC = np.log10(dataMC)
MCall = np.transpose(np.array(np.log10(dataMC))).flatten()
data_logMH = np.log10(dataMH)
MHall = np.transpose(np.array(np.log10(dataMH))).flatten()
data_logFC = np.log10(dataFC)
FCall = np.transpose(np.array(np.log10(dataFC))).flatten()
data_logFH = np.log10(dataFH)
FHall = np.transpose(np.array(np.log10(dataFH))).flatten()
```

```python
# Please Note that this data is applicable for the case of Brownian Motion
# load the dataset for IFCT: Localized (DC ~ M^0) and Systemic (DC ~ M^2/3) infection
#LN vol scaling as M^2/3
dataFC_qp = pd.read_csv('Data/ForReviewer_M^0.67_Aug20/BM_IFCT_DCConstant.csv', header = None, delimiter=',')
dataFH_qp = pd.read_csv('Data/ForReviewer_M^0.67_Aug20/BM_IFCT_DCHalf.csv', header = None,delimiter=',')

data_logFC_qp = np.log10(dataFC_qp)
FCall_qp = np.transpose(np.array(np.log10(dataFC_qp))).flatten()
data_logFH_qp = np.log10(dataFH_qp)
FHall_qp = np.transpose(np.array(np.log10(dataFH_qp))).flatten()
```

```python
# Arranging X data
mass = np.arange(1, 6, 1)
mass = np.power(10, mass)
x = np.log10(mass)
n = np.size(x, 0)

def arrangeXdata(data, x):
    n_per_mass = data.shape[0]  # should be number of simulation, ex 20
    mAllP = np.repeat(x, n_per_mass)  # length 5*20 = 100
    M = np.linspace(min(x), max(x), num=100, endpoint=True)
    return mAllP, M
mAllP_qp, M_qp = arrangeXdata(data_logFH_qp, x)
mAllP, M = arrangeXdata(data_logFH, x)
print(x)
```

```python
def predict_time(body_mass, popt, pred_exp, wORwoLog):   
    logM = np.log10(body_mass)
    if wORwoLog == "woLog":
        n1,n2 = popt
        logy = objective_withoutLog(logM, n1, n2)
        logy_pred = objective_withoutLog(logM, n1, pred_exp)
    else:
        n1,n2, n3 = popt
        logy = objective(logM, n1, n2, n3)
        logy_pred = objective(logM, n1, pred_exp, n3)
    return (10 ** logy) / 60.0, (10 ** logy_pred) / 60.0  # in minutes
```

```python
# curve fit for mean, DC ~ constant
def curve_fit_objective(mAllP, MscalingAll, M, PredictedExponent, wORwoLog): #wORwoLog = "wLog" (with Log) or "woLog" (without Log) 
    valid = ~(np.isnan(mAllP) | np.isnan(MscalingAll))
    if wORwoLog == "wLog":
        popt, covar = curve_fit(objective, mAllP, MscalingAll[valid], bounds=((-np.Inf, -np.Inf, 0),(np.Inf, np.Inf, np.Inf)))
    else:
        popt, covar = curve_fit(objective_withoutLog, mAllP, MscalingAll[valid], bounds=((-np.Inf, -np.Inf),(np.Inf, np.Inf)))

    # summarize the parameter values
    n = len(MscalingAll)    # number of data points
    alpha = 0.05 # 95% confidence interval = 100*(1-alpha)
    p = len(popt) # number of parameters
    dof = max(0, n - p) # number of degrees of freedom

    # student-t value for the dof and confidence level
    tval = t.ppf(1.0-alpha/2., dof) 
    var = np.diag(covar)
    sigma = var[0]**0.5
    n1_CI = sigma*tval
    sigma = var[1]**0.5
    n2_CI = sigma*tval

    if wORwoLog == "wLog":
        n1, n2, n3 = popt
        sigma = var[2]**0.5
        n3_CI = sigma*tval

        # calculate the output for the range M
        y = objective(M, n1, n2, n3)
        y_Upper = objective(M, n1 + n1_CI, n2 + n2_CI, n3 + n3_CI)
        y_Lower = objective(M, n1 - n1_CI, n2 - n2_CI, n3 - n3_CI)
        y_Predicted = n1 + PredictedExponent * M + np.log10((n3) + (M*np.log(10)))
        print('y = %.2f * M^{%0.2f +/- %0.2f} * log( %.2f * M)' % (n1, n2,n2_CI,n3))
    else:
        n1, n2 = popt
        # calculate the output for the range M
        y = objective_withoutLog(M, n1, n2)
        y_Upper = objective_withoutLog(M, n1 + n1_CI, n2 + n2_CI)
        y_Lower = objective_withoutLog(M, n1 - n1_CI, n2 - n2_CI)
        y_Predicted = n1 + PredictedExponent * M 
        print('y = %.2f * M^{%0.2f +/- %0.2f}' % (n1, n2,n2_CI))
        

    mice_fit,   mice_pred   = predict_time(24, popt, PredictedExponent, wORwoLog)
    kg1_fit,    kg1_pred    = predict_time(1000, popt, PredictedExponent, wORwoLog)
    human_fit,  human_pred  = predict_time(62000, popt, PredictedExponent, wORwoLog)

    print(f'mice (24 g): fit = {mice_fit:.3f} min, predicted = {mice_pred:.3f} min')
    print(f'1 kg:        fit = {kg1_fit:.3f} min, predicted = {kg1_pred:.3f} min')
    print(f'human 62 kg: fit = {human_fit:.3f} min, predicted = {human_pred:.3f} min\n')

       
    return y, y_Upper, y_Lower, y_Predicted
```

```python
def curve_fit_objective_withoutLog(mAllP, MscalingAll, M, PredictedExponent):
    valid = ~(np.isnan(mAllP) | np.isnan(MscalingAll))
    popt, covar = curve_fit(objective_withoutLog, mAllP, MscalingAll[valid], bounds=((-np.Inf, -np.Inf),(np.Inf, np.Inf)))


    # summarize the parameter values
    n = len(MscalingAll) # number of data points
    n1, n2 = popt
    sigma_ab = np.sqrt(np.diagonal(covar))

    alpha = 0.05 # 95% confidence interval = 100*(1-alpha)
    p = len(popt) # number of parameters
    dof = max(0, n - p) # number of degrees of freedom

    # student-t value for the dof and confidence level
    tval = t.ppf(1.0-alpha/2., dof) 
    var = np.diag(covar)
    sigma = var[0]**0.5
    n1_CI = sigma*tval
    sigma = var[1]**0.5
    n2_CI = sigma*tval

    print('y = %.2f * M^{%0.2f +/- %0.2f}' % (n1, n2,n2_CI))
    print('mice = %f days' %((np.power(10, n1 + n2 * np.log10(24)))/60/60/24))
    print('kg1 = %f days' %((np.power(10, n1 + n2 * np.log10(1000)))/60/60/24))
    print('Human = %f wk' %((np.power(10, n1 + n2 * np.log10(62000)))/60/60/24/7))

    # calculate the output for the range M
    y = objective_withoutLog(M, n1, n2)
    y_Upper = objective_withoutLog(M, n1 + n1_CI, n2 + n2_CI)
    y_Lower = objective_withoutLog(M, n1 - n1_CI, n2 - n2_CI)

    y_Predicted = n1 + PredictedExponent * M 
    return y, y_Upper, y_Lower, y_Predicted
```

```python
#Finding errorBar
def errorBar(data_log):
    stderror= np.std(data_log) / np.sqrt( len( data_log))
    er_Upper = np.mean(data_log) + 1.96 * stderror
    er_Lower = np.mean(data_log) - 1.96 * stderror
    return er_Upper, er_Lower
```

```python
def puttingAnImage(axes):
    image = plt.imread("LNCube.png")
    axin = axes.inset_axes([2,6.4,3.7,3.7],transform=axes.transData)    # create new inset axes in data coordinates
    axin.imshow(image)
    axin.axis('off')
```

```python
# curve fit for MFCT, DC constant
print("Curve Fitting for MFCT, DC constant")
y_MC, y_UpperMC, y_LowerMC, y_PerelsonMC = curve_fit_objective(mAllP, MCall, M, 0.5, "wLog")
er_UpperMC, er_LowerMC = errorBar(data_logMC)

# curve fit for IFCT, DC constant
print("Curve Fitting for IFCT, DC constant")
y_FC, y_UpperFC, y_LowerFC, y_PerelsonFC = curve_fit_objective(mAllP, FCall, M, 0, "wLog")
er_UpperFC, er_LowerFC = errorBar(data_logFC)

# curve fit for MFCT, DC~M^{half}
print("Curve Fitting for MFCT, DC~M^0.5")
y_MH, er_UpperMH, er_LowerMH, y_PerelsonMH = curve_fit_objective(mAllP, MHall, M, 0, "wLog")
er_UpperMH, er_LowerMH = errorBar(data_logMH)

# curve fit for IFCT, DC~M^{half}
print("Curve Fitting for IFCT, DC~M^0.5")
y_FH, er_UpperFH, er_LowerFH, y_PerelsonFH = curve_fit_objective(mAllP, FHall, M, -0.5, "wLog")
er_UpperFH, er_LowerFH = errorBar(data_logFH)
```

```python
# curve fit for IFCT, pred: IFCT~M^{0}, DC constant
y_fit_FC_qp, y_UpperFC_qp, y_LowerFC_qp, y_pred_FC_qp = curve_fit_objective(mAllP_qp, FCall_qp, M, 0, "woLog")
er_UpperFC, er_LowerFC = errorBar(data_logFC_qp)

# curve fit for IFCT, pred: IFCT~M^{-2/3}, DC~M^{-2/3}
y_fit_FH_qp, y_UpperFH_qp, y_LowerFH_qp, y_pred_FH_qp = curve_fit_objective(mAllP_qp, FHall_qp, M, -2/3, "woLog")
er_UpperFH, er_LowerFH = errorBar(data_logFH_qp)
```
